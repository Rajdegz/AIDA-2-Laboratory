# Task 4

Find the variance and standard deviation of the following scores on an exam:
92, 95, 85, 80, 75, 50

```python
task4 = sorted([92, 95, 85, 80, 75, 50])
```

```python
# Method 1
np.var(task4)
```

```python
# Method 2
def variance(data):
  n = len(data)
  mean = sum(data) / n
  deviations = [(x - mean) ** 2 for x in data]
  variance = sum(deviations) / n
  return variance

variance([92, 95, 85, 80, 75, 50])
```
